```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ecommerce_cleaned.csv')
```

```python
print(f"Total Revenue         : ${df['total_purchase_amount'].sum():,.0f}")
print(f"Total Orders          : {len(df):,}")
print(f"Unique Customers      : {df['customer_id'].nunique():,}")
print(f"Avg Order Value       : ${df['total_purchase_amount'].mean():,.2f}")
print(f"Churn Rate            : {(df['churn']=='Churned').mean()*100:.2f}%")
print(f"Return Rate           : {(df['returns']=='Yes').mean()*100:.2f}%")
print(f"Top Category          : {df.groupby('product_category')['total_purchase_amount'].sum().idxmax()}")
print(f"Top Payment Method    : {df['payment_method'].value_counts().idxmax()}")
print(f"Peak Revenue Year     : {df.groupby('purchase_year')['total_purchase_amount'].sum().idxmax()}")
print(f"Avg Age (Churned)     : {df[df['churn']=='Churned']['customer_age'].mean():.1f}")
```

```python
tableau_df = df[[
    'customer_id', 'purchase_date', 'purchase_year',
    'purchase_month', 'purchase_quarter', 'purchase_dow',
    'product_category', 'product_price', 'quantity',
    'total_purchase_amount', 'payment_method',
    'customer_age', 'age_group', 'gender',
    'returns', 'churn', 'spend_segment'
]].copy()
```

```python
tableau_df.to_csv('/content/ecommerce_tableau_ready.csv', index=False)
print(f"\nTableau file saved")
print(f"Shape   : {tableau_df.shape}")
print(f"Columns : {tableau_df.columns.tolist()}")
```
