```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/ecommerce_customer_data_custom_ratios.csv')
print("Raw shape:", df.shape)
```

```python
df.drop(columns=['Customer Name', 'Age'], inplace=True)

print(f"  Shape now: {df.shape}")
```

```python
df.columns = [
    'customer_id', 'purchase_date', 'product_category',
    'product_price', 'quantity', 'total_purchase_amount',
    'payment_method', 'customer_age', 'returns',
    'gender', 'churn'
]
print(f"  Columns: {df.columns.tolist()}")
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df['returns']       = df['returns'].astype('float')
print(df.dtypes)
```

```python
before = df['returns'].isnull().sum()
df['returns'] = df['returns'].fillna(0)
after = df['returns'].isnull().sum()
print(f"  Before: {before:,} nulls → After: {after} nulls")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['product_category'] = df['product_category'].str.strip().str.title()
df['payment_method']   = df['payment_method'].str.strip().str.title()
df['gender']           = df['gender'].str.strip().str.title()
```

```python
before = len(df)
df.drop_duplicates(inplace=True)
print(f"Duplicates removed: {before - len(df)}")
```

```python
df['product_category'] = df['product_category'].str.strip().str.title()
df['payment_method']   = df['payment_method'].str.strip().str.title()
df['gender']           = df['gender'].str.strip().str.title()
print(f"Text columns standardized")
```

```python
print(f"  Returns : {df['returns'].unique()}")
print(f"  Churn   : {df['churn'].unique()}")
```

```python
df['purchase_year']    = df['purchase_date'].dt.year
df['purchase_month']   = df['purchase_date'].dt.month
df['purchase_quarter'] = df['purchase_date'].dt.quarter
df['purchase_dow']     = df['purchase_date'].dt.day_name()
```

```python
df['age_group'] = pd.cut(
    df['customer_age'],
    bins=[17, 25, 35, 45, 55, 70],
    labels=['18–25', '26–35', '36–45', '46–55', '56–70']
)
```

```python
df['spend_segment'] = pd.cut(
    df['total_purchase_amount'],
    bins=[0, 500, 1500, 3000, 6000],
    labels=['Low', 'Medium', 'High', 'Premium']
)
```

```python
df['price_per_unit'] = (df['total_purchase_amount'] / df['quantity']).round(2)

print(f"  purchase_year, purchase_month, purchase_quarter,")
print(f"  purchase_dow, age_group, spend_segment, price_per_unit")
```

```python
print(f"  FINAL SHAPE     : {df.shape}")
print(f"  REMAINING NULLS : {df.isnull().sum().sum()}")
print(f"  DUPLICATES      : {df.duplicated().sum()}")
print(df.head(3))
```

```python
df.to_csv('/content/ecommerce_cleaned.csv', index=False)
```
